```python
def trans(a,b):
    tmp=str(a)+'_'+str(b)
    return tmp
#province	city
res['riqi'] = res.apply(lambda x: trans(x['month'],x['day']), axis=1)
```

```python
def trans(a,b):
    tmp=str(a)+'_'+str(b)
    return tmp
#province	city
res['number'] = res.apply(lambda x: trans(x['month'],x['day']/50), axis=1)
```

```python
res_sort = res.sort_values(by='number', ascending=True)
res_sort.head()
```
